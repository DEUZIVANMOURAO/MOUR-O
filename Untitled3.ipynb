```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab.patches import cv2
from matplotlib import pyplot as plt

#Ler a imagem

imagem = cv2.imread('lampada-Incandescente.png')
cv2_imshow(imagem)

print("Altura (height): %d pixels" % (imagem.shape[0]))
print ("Largura (width): %d pixels" % (imagem.shape[1]))
print ("Canais (channels): %d"    % (imagem.shape[2]))


altura, largura = imagem.shape[:2]
```

```python
#Alterando os pixels da imagem
(b, g, r) = imagem[0, 0]
print ("Cor do pixel em (0, 0) - Vermelho: %d, Verde: %d, Azul: %d" % (r, g, b))

(b, g, r) = imagem[305, 250]
print ("Cor do pixel em (250, 305) - Vermelho: %d, Verde: %d, Azul: %d" % (r, g, b))

(b, g, r) = imagem[30, 250]
print ("Cor do pixel em (250, 30) - Vermelho: %d, Verde: %d, Azul: %d" % (r, g, b))

cv2.waitKey(0)
```

```python
# Alterar a cor do pixel
imagem[300, 250] = 255
cv2_imshow(imagem)
```

```python
#Alterando uma imagem

imagem[50:150, 100:25] = (200, 240, 255)
cv2_imshow(imagem)

cv2.waitKey(0)
```

```python
#Fatiando a imagem e salvando o pedao
fatia = imagem[0:150, 150:300]
cv2_imshow(fatia)
cv2.waitKey(0)
```
