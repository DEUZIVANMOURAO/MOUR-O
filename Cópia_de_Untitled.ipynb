```python
import numpy as np
from matplotlib import pyplot as plt

x = [0,10,10,2,2,8,8,4,4,6,6,5,5]
y = [0,0,10,10,2,2,8,8,4,4,6,6,5]

plt.plot(x, y,'r')
plt.title("plot line in Matplotlib",fontsize=15)
plt.xlabel("x", fontsize=13)
plt.xlabel("y", fontsize=13)
plt.show()






















```

```python
import matplotlib.pyplot as plt

figure, axes = plt.subplots()
Drawing_uncolored_circle = plt.Circle( (0.5, 0.5, ),
                                      0.3 ,
                                       fill = False)
Drawing_uncolored_circle2 = plt.Circle( (0.5, 0.5, ),
                                      0.1 ,
                                       fill = False)
axes.set_aspect(1)
axes.add_artist( Drawing_uncolored_circle )
axes.add_artist( Drawing_uncolored_circle2 )
plt.title( 'Circle' )
plt.show()







```
