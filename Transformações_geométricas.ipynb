```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab.patches import cv2
from matplotlib import pyplot as plt

#Ler a imagem

path = ('lampada.png')
imagem = cv2.imread(path)
cv2_imshow(imagem)

print("Altura (height): %d pixels" % (imagem.shape[0]))
print ("Largura (width): %d pixels" % (imagem.shape[1]))
print ("Canais (channels): %d"    % (imagem.shape[2]))


altura, largura = imagem.shape[:2]


```

```python
#Alterando os pixels da imagem
(b, g, r) = imagem[0, 0]
print ("Cor do pixel em (0, 0) - Vermelho: %d, Verde: %d, Azul: %d" % (r, g, b))

(b, g, r) = imagem[305, 250]
print ("Cor do pixel em (250, 305) - Vermelho: %d, Verde: %d, Azul: %d" % (r, g, b))

(b, g, r) = imagem[30, 250]
print ("Cor do pixel em (250, 30) - Vermelho: %d, Verde: %d, Azul: %d" % (r, g, b))

cv2.waitKey(0)
```

```python
## Alterar a cor do pixel
imagem[305, 250] = 255
cv2_imshow(imagem)
```

```python
#Alterando uma imagem

imagem[200:350, 100:250] = (200, 240, 255)
cv2_imshow(imagem)

cv2.waitKey(0)
```

```python
#Fatiando a imagem e salvando o pedaço
fatia = imagem[0:150, 150:300]
cv2_imshow(fatia)
cv2.waitKey(0)
```

```python
#Redimensionando a imagem
cv2_imshow(imagem)
resize = cv2.resize(imagem, (200,250))
cv2_imshow( resize) # Resized Image
cv2.waitKey(0)
```

```python
def bgr2rbg(img):
    return img[:,:,::-1]

#vira a imagem horizontal
flip_h = imagem [:, :: - 1]

#Vira a imagem verticalmente
flip_v = imagem [:: - 1]

##Plota a imagem modificada
plt.subplot(221)
plt.title('Horizontally')
plt.imshow(bgr2rbg(flip_h))

plt.subplot(222)
plt.title('Vertically')
plt.imshow(bgr2rbg(flip_v))

plt.show()
```

```python
import numpy as np
#Translação (deslocamento)
deslocamento = np.float32([[1, 0, 25], [0, 1, 50]])
deslocado = cv2.warpAffine(imagem, deslocamento, (altura, largura))
cv2_imshow(deslocado)

cv2.waitKey(0)
```

```python
deslocamento = np.float32([[1, 0, -50], [0, 1, -90]])
deslocado = cv2.warpAffine(imagem, deslocamento, (largura, altura))
cv2_imshow(deslocado)

cv2.waitKey(0)
```

```python
#Rotação da imagem
ponto = (largura / 2, altura / 2) #ponto no centro da figura
rotacao = cv2.getRotationMatrix2D(ponto, 30, 1.0)
rotacionado = cv2.warpAffine(imagem, rotacao, (largura, altura))
cv2_imshow( rotacionado)

cv2.waitKey(0)
```

**Atividade de Multimídia**

Configure os parametros o algoritmo referente a rotação e translação para ajustar a imagem do sodoku. Deixe a saída como a imagem a seguir:

sodoku_out.png

```python
##Atividade

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('/content/sodoku.jpg')

rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[100,350]])
pts2 = np.float32([[10,100],[200,50],[200,300]])

M = cv.getAffineTransform(pts1,pts2)

dst = cv.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
